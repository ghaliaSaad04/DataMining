
## **Normalization**

To ensure all numerical attributes contribute proportionally to the analysis and modeling, Min-Max Scaling was applied to rescale selected features between **0 and 1**. The following columns were normalized:

- **Weather**  
- **Road_Type**  
- **Time_of_Day**  
- **Traffic_Density**  
- **Speed_Limit**  
- **Number_of_Vehicles**  
- **Driver_Alcohol**  
- **Road_Condition**  
- **Vehicle_Type**  
- **Driver_Age**  
- **Driver_Experience**  
- **Road_Light_Condition**  

The normalization process involved stripping spaces from column names, checking for missing columns, and only applying scaling to available ones. A **Min-Max Scaler** was then fitted to these columns, transforming their values to a range of **0 to 1**. This ensured consistency across attributes, improved the performance of machine learning models, and prepared the dataset for more effective clustering and classification.  

After normalization, the transformed dataset was updated in **Traffic_DataFrame**, ready for further analysis.



```python
from sklearn.preprocessing import MinMaxScaler

# Select columns to normalize
columns_to_normalize = ["Weather", "Road_Type", "Time_of_Day", "Traffic_Density", "Speed_Limit",
                        "Number_of_Vehicles", "Driver_Alcohol", "Road_Condition", "Vehicle_Type",
                        "Driver_Age", "Driver_Experience", "Road_Light_Condition"]

# Ensure column names are stripped of spaces
Traffic_DataFrame.columns = Traffic_DataFrame.columns.str.strip()

# Check for missing columns
missing_columns = [col for col in columns_to_normalize if col not in Traffic_DataFrame.columns]
if missing_columns:
    print("Warning: The following columns are missing and will be skipped:", missing_columns)

# Filter available columns
columns_to_normalize = [col for col in columns_to_normalize if col in Traffic_DataFrame.columns]

# Apply Min-Max Scaling
scaler = MinMaxScaler()
Traffic_DataFrame[columns_to_normalize] = scaler.fit_transform(Traffic_DataFrame[columns_to_normalize])

print(Traffic_DataFrame)
print("Normalization complete!")
```

## **Discretization**

In this process, we transformed continuous numerical values into categories by splitting them into three groups (bins). First, we identified all the numerical columns in the dataset and then applied equal-width binning to divide their values into discrete categories.

Discretization helps make the data easier to analyze by grouping similar values together. This can improve the performance of certain machine learning models, reduce the impact of outliers, and make patterns in the data more noticeable.




```python
# Number of bins for discretization
num_bins = 3  

# Automatically find numerical columns
numeric_columns = Traffic_DataFrame.select_dtypes(include=['number']).columns  

# Discretize all numerical columns
for col in numeric_columns:
    Traffic_DataFrame['discretized_' + col] = pd.cut(Traffic_DataFrame[col], bins=num_bins, labels=False)

# Display the transformed DataFrame
print("Original and Discretized DataFrame:")
print(Traffic_DataFrame)
```

## **Feature selection**
Using the encoded dataset, we 
applied feature selection to identify the most significant factors influencing accident occurrence. By employing statistical methods to evaluate the correlation between features and the target variable (**Accident**), we identified key attributes that contribute to road accidents.  

The top selected features include Road_Type (which determines the complexity and risk level of the driving environment), Time_of_Day (as traffic patterns and visibility conditions vary throughout the day), Road_Condition (which impacts vehicle control and safety), Driver_Age (a crucial factor influencing driving experience and reaction time), and Driver_Experience (which directly affects decision-making and risk assessment on the road).  

Using a univariate feature selection approach (SelectKBest with f_classif), we ranked the top five features based on their statistical significance. This selection process enhances model efficiency by reducing dimensionality while preserving the most relevant predictors, ultimately improving accident risk assessment and prediction accuracy.

```python
import pandas as pd
from sklearn.feature_selection import SelectKBest, f_classif

# Separate features and target variable
X = Traffic_DataFrame.drop(columns=["Accident"])  # Features
y = Traffic_DataFrame["Accident"]  # Target variable

# Apply SelectKBest to select the top 5 features
selector = SelectKBest(score_func=f_classif, k=5)
X_new = selector.fit_transform(X, y)

# Get the selected feature names
selected_features = X.columns[selector.get_support()]
print("Top selected features:", selected_features)
```
