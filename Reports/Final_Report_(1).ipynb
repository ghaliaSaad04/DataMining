## **Removing missing values**:
We decided to separate categorical data and numerical data for easy handling, Categorical columns were filled with their mode, assuming the most common condition represents missing cases.Numerical columns were filled with median, to avoid skewing by outliers.

```python
import pandas as pd

# **Handle Missing Values**

# Separate categorical and numerical columns
categorical_cols = Traffic_DataFrame.select_dtypes(include='object').columns
numerical_cols = Traffic_DataFrame.select_dtypes(include='float64').columns

# Fill missing values for categorical columns
for col in categorical_cols:
    Traffic_DataFrame[col] = Traffic_DataFrame[col].fillna(Traffic_DataFrame[col].mode()[0])

# Fill missing values for numerical columns
for col in numerical_cols:
    Traffic_DataFrame[col] = Traffic_DataFrame[col].fillna(Traffic_DataFrame[col].median())

display(Traffic_DataFrame)
print("-"*40)
Traffic_DataFrame.info()
count = pd.isnull(Traffic_DataFrame).sum().sum()
print(f"The Number Of Null Values Is: {count}")
```

We Applied the Interquartile Range (IQR) method to remove values that fall far outside typical ranges, then used box plots to show the result

```python
# *** Outlier Removal using IQR ***
def remove_outliers_iqr(Traffic_DataFrame, columns):
    for col in columns:
        Q1 = Traffic_DataFrame[col].quantile(0.25)
        Q3 = Traffic_DataFrame[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        Traffic_DataFrame = Traffic_DataFrame[(Traffic_DataFrame[col] >= lower_bound) & (Traffic_DataFrame[col] <= upper_bound)]
    return Traffic_DataFrame
Traffic_DataFrame = remove_outliers_iqr(Traffic_DataFrame, numerical_cols)

#Box plot
import matplotlib.pyplot as plt
import seaborn as sns

numerical_cols = Traffic_DataFrame.select_dtypes(include='float64').columns

fig, axes = plt.subplots(len(numerical_cols), 1, figsize=(12, 4 * len(numerical_cols)))

for i, col in enumerate(numerical_cols):
    sns.boxplot(x=Traffic_DataFrame[col], ax=axes[i], color='skyblue')
    axes[i].set_title(f'Boxplot: {col}', fontsize=14)
    axes[i].set_xlabel('')
    axes[i].grid(True)

plt.tight_layout()
plt.show()
```
